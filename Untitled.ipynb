```python
import data_parse as dp
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
import json
```

```python
with open('data/recipes_raw_nosource_ar.json') as f:
    data = json.load(f)
ingredient_lists = [] 
for recipe in data:
    if 'ingredients' in data[recipe]:
        ingredient_lists.append(np.array(dp.extract_ingredients(data[recipe]['ingredients'])))

        
```

```python
from sklearn.feature_extraction.text import CountVectorizer
def my_preprocessor(doc):
	return doc

def my_tokenizer(x):
	return x

def create_vectorization(training_X):
	vectorizer  = CountVectorizer(preprocessor = my_preprocessor, tokenizer = my_tokenizer)
	vectorizer.fit_transform(training_X)
	return vectorizer

v = create_vectorization(ingredient_lists)
print (v.get_feature_names())
print( v.vocabulary_)

# print(v.transform(li))
```

```python
len(ingredient_lists)-len(v.vocabulary_)
```

```python
from sklearn.model_selection import train_test_split
data = np.array(ingredient_lists)
print(len(data))
train_idx, val_idx = train_test_split(data, test_size =0.8)
```

```python
transform = v.transform(train_idx)
transform.shape
```

```python
import random
labels = np.zeros(len(train_idx))
data_train = transform
# print(data_train)
for k in range(len(train_idx)):
#     for j in range(1):
    word_idx = v.vocabulary_[str(random.choice(train_idx[k]))]
    labels[k] = word_idx
    data_train[k,word_idx] = 0
```

```python
data_train.shape
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self, vocab_size):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(vocab_size, 120)
        self.fc2 =  nn.Linear(120, vocab_size)
        self.vocab_size = vocab_size
    def forward(self, x):
        out = F.relu(self.fc1(x))
        out = F.softmax(self.fc2(out))
        return out
#         self.fc3 = nn.Linearc
```

```python
data_coo = data_train.tocoo()
data_coo.shape
```

```python
import torch
print(len(data_coo.col.tolist()))
data_torch = torch.sparse.FloatTensor(torch.LongTensor([data_coo.row.tolist(), data_coo.col.tolist()]),
                              torch.FloatTensor(data_coo.data.astype(np.int64)),data_coo.shape)
```

```python
net = Net(len(v.vocabulary_))
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
for epoch in range(2):  # loop over the dataset multiple times

    running_loss = 0.0
#     for i, data in enumerate(data_torch, 0):
#         print(torch.Tensor(labels[i]).shape)
        # get the inputs; data is a list of [inputs, labels]
    inputs, label = data_torch,torch.Tensor(labels).type(torch.LongTensor)

# zero the parameter gradients
    optimizer.zero_grad()

    # forward + backward + optimize
    outputs = net(inputs)
    print(outputs.shape)
    print(label.shape)
    print(label)
    loss = criterion(outputs, label)
    loss.backward()
    optimizer.step()

    # print statistics
    running_loss += loss.item()
#     if i % 2000 == 1999:    # print every 2000 mini-batches
    print('[%d, %5d] loss: %.3f' %
          (epoch + 1,  1, running_loss / 2000))
    running_loss = 0.0

print('Finished Training')
```
